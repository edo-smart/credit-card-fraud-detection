Step 10: Target Variable Distribution

## Observation

The dataset is highly imbalanced.

Legitimate transactions represent approximately **99.83%** of the dataset, while fraudulent transactions account for only about **0.17%**.

This severe imbalance means that accuracy alone is not an appropriate evaluation metric.

```python
df["Class"].value_counts().plot(kind="bar")

plt.title("Fraud vs Legitimate Transactions")
plt.xlabel("Class")
plt.ylabel("Count")

plt.show()
```

Step 11: Visualize the Target

## Observation

The bar chart clearly illustrates the class imbalance.

The overwhelming majority of transactions are legitimate, while fraudulent transactions make up only a very small portion of the dataset.

Special techniques such as resampling or class weighting may be required during model development.
